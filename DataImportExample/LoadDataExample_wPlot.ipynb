```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('BOII_SR_2000to2017.csv', names=['Year','Month','Day','SolarRad_MJd'], skiprows=1); # Import csv file, skipping and naming header
df[0:5] # check first 5 records in data frame
```

```python
df['Date'] = pd.to_datetime(df.Year*10000+df.Month*100+df.Day,format='%Y%m%d') # create better date format
df[0:5] # check first 5 records in frame
```

```python
df.plot(x='Date', y='SolarRad_MJd',figsize=(14,10))
plt.show()
```

```python
# I'm leaving your comments for this section Lejo
# They do a good job of explaining what is going on

# Use the pandas .values operator to load the values of discharge (the 'Q' column)
# from the pandas dataframe into a numpy array
SR = df['SolarRad_MJd'].values # 

# Use numpy array indexing to get discharge from October 2, 2000 to September 30, 2017. 
# The 'end' of a numpy array can be indexed using -1: as if the index in the negative direction
# wraps back around to the end of the record
SRt   = SR[1:-1] 

# Similarly get solar radiation from October 1, 2000 to September 29, 2017 by getting data from the 
# beginning of the record to the 'end' minus 1 using the -2 as the index. 
SRtm1 = SR[0:-2]

# Use Matplotlib to create a scatter plot of the two time series, with a title and labeled axes
plt.figure(figsize=(14,10))
plt.plot(SRtm1,SRt,'o')
plt.title('How correlated is today\'s solar radiation with yesterday\'s?')
plt.xlabel('$SR_{t-1}$ MJ / m$^2$ / day')
plt.ylabel('$SR_t$ MJ / m$^2$ / day')
plt.show()
```
